```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import re
pd.set_option('display.max_columns', None)
```

```python
def url_access(url_page):
    resp_codes=requests.get(url_page)
    if resp_codes.status_code == 200:
        print('Accesso correcto')
        sopa_codes=BeautifulSoup(resp_codes.content, 'html.parser')
        return sopa_codes
    else:
        print('Error requests')
        print(resp_codes)
        
```

```python

sopa_language=url_access(url_page='https://es.wikipedia.org/wiki/ISO_639-1')
```

```python
tablas=sopa_language.find_all('table')
```

```python
language_tabla=tablas[1].find_all("td")
```

```python
resultado=[]
for fila in language_tabla:
    fila_texto=fila.text
    elementos_fila=fila_texto.split("\n")
    resultado.append(elementos_fila)

```

```python
patron='\\[a-z]'
resultado=[]
for fila in language_tabla:
    fila_texto=re.sub(patron,"",fila.text)
    resultado.append(fila_texto)
```
