## Parte 2

Vamos a extraer los datos de un libro: Romeo and Juliet


```python
#Estraemos un archivo .txt
rddRomeoJulieta=sc.textFile("romeo_y_julieta.txt")
```

```python
#Comprobamos que que se haya creado correctamente el rdd
rddRomeoJulieta
```

```python
#PAra obtener el libro entero y comprobar que nuestro objeto rdd funciona correctamente:
rddRomeoJulieta.collect()
```

```python
#Separaremos por espacios para obtener todas las palabras del libro:
rdd2RomeoJulieta=rddRomeoJulieta.flatMap(lambda x: x.split(" "))
```

```python
rdd2RomeoJulieta.collect()
```

```python
rdd2RomeoJulieta.getNumPartitions()
```

```python
#Obtenemos los datos de la primera particion que corresponde a la  mitad del libro
rdd2RomeoJulieta.glom().collect()[0]
```

```python
#Contamos el numero de elementos:
rdd2RomeoJulieta.count()
```

```python
rdd3RomeoJulieta=rdd2RomeoJulieta.map(lambda x: x.upper()) #Pasamos cada palabra a mayúscula
rdd3RomeoJulieta.take(10) # con .take(n), cogemos n elementos.
```

```python
rdd4RomeoJulieta=rdd3RomeoJulieta.filter(lambda x:x.startswith("W"))#Filtramos las palabras que empiezan por "W"
rdd4RomeoJulieta.take(1)
```

```python
#contaremos cuantas veces se repite cada palabra
rddCount=rdd2RomeoJulieta.map(lambda x: (x,1)) #Añadimos un 1 a cad apalabra en forma de tupla.
#Esta nos indicará el número de veces que se repite cada palabra.
rddCount.take(5)
```

```python
rddCount2=rddCount.reduceByKey(lambda a,b: a+b) # .reduceByKeys, cada palabra o key se reduce cada vez que se repite y sumamos cada vez que se repite cada palabra .
rddCount2.take(5)
```

```python
spark
```

```python
#Creamos un dataframe:
from datetime import  datetime, date
import pandas as pd
from pyspark.sql import Row
```

```python
df= spark.createDataFrame([
    (1,2.,'Alberto',date(2030,1,1), datetime(2030,1,1,12,0)),
    (2,2.,'Marta',date(2040,1,1), datetime(2040,1,1,7,0)),
    (3,2.,'Pablo',date(2050,1,1), datetime(2050,1,1,5,0)),
], schema='id long, note double, name string, date date, datestamp timestamp'
)
df.show()
```

```python
df.rdd.getNumPartitions()
```

```python
#Tenemos 8 particiones pero tres filas:
df.rdd.glom().collect()
```

```python
#Observamos que las particiones 0,1,3,4,6 están vacías.
df.printSchema()
```

```python
spark.conf.set('spark.sql.repl.eagerEval.enabled', True)
```
